```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import openpyxl
import seaborn as sns 
```

```python
df = pd.read_excel('C:/Users/Admin/Downloads/Employee Sample Data - Copy.xlsx')
```

```python
df.dropna(inplace = True)
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=15, edgecolor='black',color='pink')
plt.title('Distribution of Employee Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Employee Gender Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
department_counts = df['Department'].value_counts()
plt.figure(figsize=(8, 6))
department_counts.plot(kind='bar', color='purple')
plt.title('Number of Employees by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
df['Ethnicity'].value_counts().plot(kind='bar',color='red')
plt.xlabel('Ethnicity')
plt.ylabel('frequency')
plt.title('Distribution Ethnicity')
plt.show()
     
```

```python
df['Department'].value_counts().plot(kind='bar',color='magenta')
plt.xlabel('Department Name')
plt.ylabel('frequency')
plt.title('Distribution Department')
plt.show()
```

```python
df['Job Title'].value_counts().plot(kind='bar',color='brown')
plt.xlabel('Job Title')
plt.ylabel('frequency')
plt.title('Job Titles')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Annual Salary'], bins=15,edgecolor='black', color='green')
plt.title(' Annual Salary Distribution')
plt.xlabel('Annual Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
sns.displot(df, x="Age", col="Gender",color='olive')

plt.show()
     
```

```python
sns.displot(df, x="Department", col="Gender",color='gold')     
plt.xticks(x="Department", rotation='vertical')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Country'], df['Annual Salary'], color='darkred')
plt.title('Distribution of country & Annual Salary')
plt.xlabel('Country')
plt.ylabel('Annual Salary')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['City'], df['Annual Salary'], color='deep')
plt.title('Distribution of city & Annual Salary')
plt.xlabel('City')
plt.xticks(x="City", rotation='vertical')
plt.ylabel('Annual Salary')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Job Title'], df['Annual Salary'], color='b')
plt.title('Distribution of Job Title & Annual Salary')
plt.xlabel('Job Title')
plt.xticks(x="Job Title", rotation='vertical')
plt.ylabel('Annual Salary')
plt.grid(True)
plt.show()
plt.xticks(x="Department", rotation='vertical')
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Ethnicity'], df['City'], color='b')
plt.title('Distribution of city & Ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('City')
plt.grid(True)
plt.show()
```
