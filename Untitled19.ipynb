```python
data.columns = data.columns.str.strip()
formula = 'Q("Price (USD)") ~ Year + Q("Engine Size (L)")'
model = smf.ols(formula=formula, data=data)
results = model.fit()
print(results.summary())
regression_table = pd.DataFrame({
    'coefficients': results.params,
    'std_errors': results.bse,
    't_values': results.tvalues,
    'p_values': results.pvalues
})
print(regression_table)
```

```python
data.columns = data.columns.str.strip()
data_cleaned = data.dropna(subset=['Price (USD)', 'Engine Size (L)'])
data_cleaned['Price (USD)'] = data_cleaned['Price (USD)'].astype(float)
data_cleaned['Engine Size (L)'] = data_cleaned['Engine Size (L)'].astype(float)
x = data_cleaned['Price (USD)']
y = data_cleaned['Engine Size (L)']
slope, intercept, r, p, std_err = stats.linregress(x, y)
def myfunc(x):
    return slope * x + intercept
mymodel = list(map(myfunc, x))
plt.scatter(x, y)
plt.plot(x, mymodel, color='red')
plt.ylim(ymin=0, ymax=10)  
plt.xlim(xmin=0, xmax=200000)  
plt.xlabel("Price (USD)")
plt.ylabel("Engine Size (L)")
plt.title("Linear Regression: Engine Size vs Price")
plt.show()
```

```python
data.columns = data.columns.str.strip()
data_filtered = data[data['Year'] != 2020]
data_filtered = data_filtered.dropna(subset=['Year', 'Price (USD)'])
data_filtered['Year'] = data_filtered['Year'].astype(int)
data_filtered['Price (USD)'] = data_filtered['Price (USD)'].astype(float)
x = data_filtered['Year']
y = data_filtered['Price (USD)']
slope, intercept, r, p, std_err = stats.linregress(x, y)
def myfunc(x):
    return slope * x + intercept
mymodel = list(map(myfunc, x))
plt.scatter(x, y, label='Data Points')
plt.plot(x, mymodel, color='red', label='Regression Line')
plt.xlabel("Year")
plt.ylabel("Price (USD)")
plt.title("Linear Regression: Price vs Year (excluding 2020)")
plt.legend()
plt.show()
```
